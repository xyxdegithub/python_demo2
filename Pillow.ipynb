```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img1 = Image.open("image/Lena.jpg")
```

```python
img1.save("image/Lena.png")
img2=Image.open("image/Lena.png")
```

```python
img1.save("image/Lena.bmp")
img3=Image.open("image/Lena.bmp")
```

```python
print(img1.format)
print(img1.size)
print(img1.mode)
print(img2.format)
print(img2.size)
print(img2.mode)
print(img3.format)
print(img3.size)
```

```python
plt.subplot(1,3,1)
plt.axis("off")
plt.title(img1.format)
plt.imshow(img1)
plt.subplot(1,3,2)
plt.axis("off")
plt.title(img2.format)
plt.imshow(img2)
plt.subplot(1,3,3)
plt.axis("off")
plt.title(img3.format)
plt.imshow(img3)
```

```python
#
img_gray=img1.convert("L")
print(img_gray.mode)
plt.imshow(img_gray)
img_gray.save("image/lena_gray.png")
```
